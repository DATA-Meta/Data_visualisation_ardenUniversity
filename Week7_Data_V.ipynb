```python
import pandas as pd
import numpy as numpy
import plotly.express as px
```

```python
df = px.data.gapminder()
df.head()
```

```python
df_2007 = df[df['year'] == 2007]

fig = px.scatter(
    df_2007,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    color='continent',
    hover_name='country',
    log_x=True,
    size_max=60,
    title='GDP per Capita vs Life Expectancy (2007)'
)

fig.show()
```

```python
#Task 2 : Add Controls with ipywidgets (20 min)

# Student Instructions:

# Add an integer slider for selecting a year.

# Add a dropdown for selecting a continent (or “All”).
# Link both controls to the chart-update function.
# When a control changes, the plot updates automatically.
# Hints

# import ipywidgets as widgets
# _
# from IPython.display import display
```

```python
import ipywidgets as widgets
from IPython.display import display

# Create output area for the plot
output = widgets.Output()

# Create widgets
year_slider = widgets.IntSlider(
    value=2007,
    min=int(df["year"].min()),
    max=int(df["year"].max()),
    step=5,
    description="Year:",
    continuous_update=False,
    style={"description_width": "initial"}
)

continent_dropdown = widgets.Dropdown(
    options=["All"] + sorted(df["continent"].unique().tolist()),
    value="All",
    description="Continent:",
    style={"description_width": "initial"}
)

# Update function
def update_plot(change=None):
    selected_year = year_slider.value
    selected_continent = continent_dropdown.value

    filtered = df[df["year"] == selected_year]
    if selected_continent != "All":
        filtered = filtered[filtered["continent"] == selected_continent]

    fig = px.scatter(
        filtered,
        x="gdpPercap",
        y="lifeExp",
        size="pop",
        color="continent",
        hover_name="country",
        log_x=True,
        size_max=60,
        title=f"GDP per Capita vs Life Expectancy ({selected_year})"
    )

    with output:
        output.clear_output(wait=True)
        fig.show()

# Link widgets to update function
year_slider.observe(update_plot, names="value")
continent_dropdown.observe(update_plot, names="value")

# Display controls and plot
display(widgets.VBox([year_slider, continent_dropdown, output]))

# Initial render
update_plot()
```
